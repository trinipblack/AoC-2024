```python
input1_path = 'inputs/input_03_1.txt'
# read the input file


with open(input1_path, 'r') as file:
    input1 = file.read()

program = input1
program[:100]
```

```python
muls = program.split('mul(')
```

```python
muls
```

```python
good_muls = []
for mul in muls:
    try:
        inside = mul.split(')')[0]
        if inside.count(',') == 1:
            good_muls.append(inside)
    except:
        continue
good_muls
```

```python
result = 0
for mul in good_muls:
    a, b = mul.split(',')[0], mul.split(',')[1]

    if len(a) <= 3 and len(b) <= 3:
        try:
            mul_result = int(a) * int(b)
            print(f'Multiplying {a} and {b} to get {mul_result}')
            result += mul_result
        except:
            continue

result
```

```python
donts = "xmul(2,4)&mul[3,7]!^don't()_mul(5,5)+mul(32,64](mul(11,8)undo()?mul(8,5))".split("don't()")
```

```python
donts
```

```python
dos = [donts[0]]
for line in donts:
    if "do()" in line:
        correct = line.split("do()")[1]
        dos.append(correct)
dos_string = ''.join(dos)
```

```python
dos_string
```

```python
muls_do = dos_string.split('mul(')
```

```python
good_muls2 = []
for mul in muls_do:
    if ")" in mul:
        inside = mul.split(')')[0]
        if inside.count(',') == 1:
            good_muls2.append(inside)
good_muls2
```

```python
result2 = 0
for mul in good_muls2:
    a, b = mul.split(',')[0], mul.split(',')[1]

    if len(a) <= 3 and len(b) <= 3:
        try:
            mul_result = int(a) * int(b)
            print(f'Multiplying {a} and {b} to get {mul_result}')
            result2 += mul_result
        except:
            continue

result2
```
