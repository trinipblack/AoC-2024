```python
input1_path = 'inputs/input_05_1.txt'
# read the input file


with open(input1_path, 'r') as file:
    input1 = file.read()
```

```python
def preprocess(ip):
    page_ord_rules, pages_in_updates  = ip.split('\n\n')
    page_ord_rules = page_ord_rules.split('\n')
    pages_in_updates = pages_in_updates.split('\n')
    if pages_in_updates[-1] == '':
        pages_in_updates = pages_in_updates[:-1]
    if page_ord_rules[-1] == '':
        page_ord_rules = page_ord_rules[:-1]
    return page_ord_rules, pages_in_updates
```

```python
def check_one_rule_one_update(update, rule):
    prev_page, post_page = rule.split('|')
    update_list = update.split(',')
    if prev_page not in update_list or post_page not in update_list:
        return True
    if update_list.index(prev_page) < update_list.index(post_page):
        return True
    return False
    
def check_all_rules_one_update(update, rules):
    for rule in rules:
        if not check_one_rule_one_update(update, rule):
            return False
    return True

def find_middle_page_of_update(update):
    update_list = update.split(',')
    middle = (len(update_list) // 2)
    middle_value = int(update_list[middle])
    return middle_value

def find_middle_page_of_all_correct_updates(updates, rules):
    counter = 0
    for update in updates:
        if check_all_rules_one_update(update, rules):
            middle_value = find_middle_page_of_update(update)
            print(f'Found correct update: {update} with middle value: {middle_value}')
            counter += middle_value
    return counter
    
    
```
