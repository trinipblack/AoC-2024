```python
input1_path = 'inputs/input_04_1.txt'
# read the input file


with open(input1_path, 'r') as file:
    input1 = file.read()

ws = input1.split('\n')[:-1]
```

```python
ws
```

```python
word = 'XMAS'
def search_horizontal(w, word = word):
    print(f'Horizontal search for {word}')
    counter = 0
    for i, line in enumerate(w):
        count = line.count(word)
        if count > 0:
            counter += count
            print(f'\tFound word {word} {count} times in line {i}')
    return counter

search_horizontal(ws)
```

```python
def search_vertical(w, word = word):
    print(f'Vertical search for {word}')
    counter = 0
    for col in range(len(w[0])):
        for row in range(len(w)):
            if w[row][col] == word[0]:
                possible_word = ''.join([line[col] for line in w[row:row+len(word)]])
                if possible_word == word:
                    counter += 1
                    print(f'\tFound word {word} starting at {row, col}')
    return counter

search_vertical(ws, 'XMAS')
```

```python
search_vertical(ws, 'SAMX')
```

```python
def search_diagonal(w, word = word):
    print(f'Diagonal for word {word}')
    counter = 0
    for row in range(len(w)):
        for col in range(len(w[0])):
            if w[row][col] == word[0]:
                try:
                    possible_word = ''.join([w[row+i][col+i] for i in range(len(word))])
                    if possible_word == word:
                        counter += 1
                        print(f'\tFound word {word} starting at {row, col}')
                except:
                    pass
    return counter

def search_anti_diagonal(w, word = word):
    print(f'Anti-diagonal for word {word}')
    counter = 0
    l1 = list(range(len(w)))
    for row in l1:
        l2 = list(range(len(w[0])))
        for col in l2:
            if w[row][col] == word[0]:
                try:
                    possible_word = ''.join([w[row-i][col+i] for i in range(len(word))])
                    if possible_word == word:
                        counter += 1
                        print(f'\tFound word {word} starting at {row, col}')
                except:
                    pass
    return counter

search_diagonal(ws, 'XMAS')
```

```python
search_diagonal(ws, 'SAMX')
```

```python
search_anti_diagonal(ws, 'XMAS')
```

```python
search_anti_diagonal(ws, 'SAMX')
```

```python

def calculate_total(inp):
    total = (
        search_horizontal(inp, 'XMAS') +
        search_horizontal(inp, 'SAMX') +
        search_vertical(inp, 'XMAS') +
        search_vertical(inp, 'SAMX') +
        search_diagonal(inp, 'XMAS') +
        search_diagonal(inp, 'SAMX') +
        search_anti_diagonal(inp, 'XMAS') +
        search_anti_diagonal(inp, 'SAMX')
    )
    return total
```
