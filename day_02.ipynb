```python
def check_if_report_is_valid(report):
    return check_if_increasing(report) or check_if_decreasing(report)

def check_if_increasing(report):
    for i in range(len(report)-1):
        if report[i] >= report[i+1]:
            return False
        elif report[i+1] - report[i] > 3:
            return False
    return True

def check_if_decreasing(report):
    for i in range(len(report)-1):
        if report[i] <= report[i+1]:
            return False
        elif report[i] - report[i+1] > 3:
            return False
    return True

def count_valid_reports(reports):
    valid_reports = 0
    for report in reports:
        report_list = [int(lev) for lev in report.split()]
        if check_if_report_is_valid(report_list):
            valid_reports += 1
    return valid_reports
```

```python
def check_if_report_is_valid_2(report):

    incr, n = check_if_increasing_2(report)
    if not incr:
        incr, _ = check_if_increasing_2(modify_report(report, n))
        if not incr:
            incr, _ = check_if_increasing_2(modify_report(report, n-1))

    decr, n = check_if_decreasing_2(report)
    if not decr:
        decr, _ = check_if_decreasing_2(modify_report(report, n))
        if not decr:
            decr, _ = check_if_decreasing_2(modify_report(report, n-1))

    return incr or decr

def modify_report(report, n):
    mod_report = report.copy()
    print(f'Removing number {mod_report[n]} from {mod_report}')
    mod_report.pop(n)
    return mod_report

def check_if_increasing_2(report):
    for i in range(len(report)-1):
        if report[i] >= report[i+1]:
            return False, i+1
        elif report[i+1] - report[i] > 3:
            return False, i+1
    return True, None

def check_if_decreasing_2(report):
    for i in range(len(report)-1):
        if report[i] <= report[i+1]:
            return False, i+1
        elif report[i] - report[i+1] > 3:
            return False, i+1
    return True, None



def count_valid_reports_2(reports):
    valid_reports = 0
    for report in reports:
        report_list = [int(lev) for lev in report.split()]
        if check_if_report_is_valid_2(report_list):
            valid_reports += 1
    return valid_reports
```
