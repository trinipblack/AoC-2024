```python
input1_path = 'inputs/input_01_1.txt'
# read the input file

with open(input1_path, 'r') as file:
    input1 = file.read()
input1[:40]
```

```python
lines = input1.split('\n')
lines
```

```python
left = [int(line.replace('   ', ':').split(':')[0]) for line in lines if '   ' in line]
right = [int(line.replace('   ', ':').split(':')[1]) for line in lines if '   ' in line]
print(len(left), len(right))
left[:5], right[:5]
```

```python
# sort the left and right lists
left.sort()
right.sort()
left[:5], right[:5]
```

```python
left[0], right[0]
```

```python
import numpy as np
difference = [np.abs(l - r) for l, r in zip(left, right)]
np.array(difference).sum()
```

```python
li = [0, 1, 1, 2, 3, 3, 3]
li.count(1)
```

```python
n_map = {}
for n in set(left):
    n_map[n] = right.count(n)
```

```python
from tqdm import tqdm
```

```python
total = 0
for n in tqdm(left, total = len(left)):
    total += n_map[n] * n
total
```
